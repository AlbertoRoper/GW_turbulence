```python
import cosmoGW
import matplotlib.pyplot as plt
import interferometry as inte
import astropy.units as u
import numpy as np
import os

CWD = os.getcwd()
os.chdir('..')
fs, LISA_Om, LISA_OmPLS, LISA_Xi, LISA_XiPLS = \
            inte.read_sens(SNR=10, T=4, interf='LISA', Xi=True)
fs_comb, LISA_Taiji_Xi, LISA_Taiji_XiPLS = \
            inte.read_sens(SNR=10, T=4, interf='comb')
os.chdir(CWD)

plt.plot(fs, LISA_Om)
plt.plot(fs, LISA_OmPLS)
T = 3e5*u.GeV
g = 106

# runsE_l, runsE_nl, colE = gen.select_runs(runs, A='E')
# runE_l = runsE_l[1]
# runE_nl = runsE_nl[1]
runE_l = runs.get('E1_l_toff')
runE_nl = runs.get('E1_nl_toff')
colE = 'purple'

#gen.plot_OmGW_f(runE_l, runE_nl, T, g, col=colE)

GW = runE_nl.spectra.get('EGW_stat')
t = runE_nl.spectra.get('t_EGW')
GW_l = runE_l.spectra.get('EGW_stat')
XiGW = runE_nl.spectra.get('helEGW_stat')
XiGW_l = runE_l.spectra.get('helEGW_stat')
# indt = np.argmin(abs(t - 1))
# GW = GW[indt, :]
# GW_l = GW_l[indt, 1:]
diffGW_l = GW - GW_l
diffXiGW_l = XiGW - XiGW_l
k_nl = runE_nl.spectra.get('k')
f_nl, OmGW_nl = cosmoGW.shift_OmGW_today(k_nl, GW*k_nl, T, g)
f_nl, XiGW_nl = cosmoGW.shift_OmGW_today(k_nl, XiGW*k_nl, T, g)

k_l = runE_l.spectra.get('k')
f_l, diffOmGW_l = cosmoGW.shift_OmGW_today(k_l, diffGW_l*k_l, T, g)
f_l, XiGW_l = cosmoGW.shift_OmGW_today(k_l, XiGW_l*k_nl, T, g)
f_l, diffXiGW_nl = cosmoGW.shift_OmGW_today(k_l, diffXiGW_l*k_l, T, g)
plt.plot(f_l[1:], diffOmGW_l[1:])
plt.plot(f_nl[1:], OmGW_nl[1:])

ind = np.argmax(OmGW_nl)
ind_l = np.argmax(diffOmGW_l)
print(f_nl[ind], f_l[ind])
LISA_Om_p = np.interp(f_nl[1].value, fs, LISA_OmPLS)
print(LISA_Om_p)

plt.plot(f_nl[1], LISA_Om_p, 'o')
plt.plot(f_nl[1], OmGW_nl[1], 'o')
plt.plot(f_l[1], diffOmGW_l[1], 'o')
print(OmGW_nl[1]/LISA_Om_p)
print(diffOmGW_l[1]/LISA_Om_p)

plt.xscale('log')
plt.yscale('log')

plt.figure()
plt.plot(fs, LISA_Xi)
plt.plot(fs, LISA_XiPLS)
plt.plot(fs_comb, LISA_Taiji_Xi)
plt.plot(f_l[1:], diffXiGW_nl[1:], color='blue')
plt.plot(f_l[1:], XiGW_nl[1:])
#plt.plot(f_l[1:], XiGW_nl[1:] - XiGW_l[1:])
plt.xscale('log')
plt.yscale('log')
```

```python
OmGW_nl = np.array(OmGW_nl, dtype='float')
diffOmGW_l = np.array(diffOmGW_l, dtype='float')

SNR_v = inte.SNR(f_nl[1:], OmGW_nl[1:], fs, LISA_Om, T=4)
SNR_v2 = inte.SNR(f_l[1:], diffOmGW_l[1:], fs, LISA_Om, T=4)

print(SNR_v, SNR_v2)
```
