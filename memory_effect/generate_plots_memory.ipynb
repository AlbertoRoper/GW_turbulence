```python
plt.figure(figsize=(12, 8))

def plot_pol(run, run_nl, leg, alpha, color='blue'):
    
    XiGW = run_nl.spectra.get('helEGW')[-1, :]
    x = range(0, len(XiGW))
    x_pos, x_neg, f_pos, f_neg, col = spectra.red_blue_func(x, XiGW)
    #plt.plot(x_pos, f_pos, '.', color='red')
    #plt.plot(x_neg, abs(f_neg), '.', color='blue')

    #plt.plot(abs(XiGW), color='red', ls='-.')
    EGW = run_nl.spectra.get('EGW')[-1, :]
    plt.plot(XiGW[np.where(EGW!=0)]/EGW[np.where(EGW!=0)],
             color=color, ls='--', label=leg, alpha=alpha)

    XiGW = run.spectra.get('helEGW')[-1, :]
    x = range(0, len(XiGW))
    x_pos, x_neg, f_pos, f_neg, col = spectra.red_blue_func(x, XiGW, col=0)
    #plt.plot(x_pos, f_pos, 'o', color='red')
    #plt.plot(x_neg, abs(f_neg), 'o', color='blue')

    #plt.plot(abs(XiGW), color='red', ls='-.')
    EGW = run.spectra.get('EGW')[-1, :]
    plt.plot(XiGW[np.where(EGW!=0)]/EGW[np.where(EGW!=0)],
             color=color, alpha=alpha)
    
plot_pol(C1_l, C1_nl, r'${\cal E}_{\rm EM} = 10$', 1)
plot_pol(C2_l, C2_nl, r'${\cal E}_{\rm EM} = 1$', .7)
plot_pol(C3_l, C3_nl, r'${\cal E}_{\rm EM} = 0.1$', .5)
plot_pol(C4_l, C4_nl, r'${\cal E}_{\rm EM} = 0.02$', .3)

plt.legend(loc='lower right', fontsize=16, frameon=False)
plt.xscale('log')
#plt.yscale('log')
plt.xlim(1, 300)
plot_sets.axes_lines()
plt.ylim(-1, 1.1)
#_ = plt.yticks(np.logspace(-18, 2, 11))
plt.xlabel('$k$')
plt.ylabel(r'${\cal P}_{\rm GW}$')
ax = plt.gca()
ax.tick_params(axis='x', pad=15)
```
